### Casting

##### int to float

```
int_var = 14
float_var = float(14)
float_var
```

##### float to int

```
float_var = 54.464
int_var = int(float_var)
int_var
```

##### numeric to string

```
num_var = 54.47561
str_var = str(num_var)
str_var
```

##### string to numeric (string number)

```
str_var = '44'
num_var = int(str_var)
num_var
```

##### string to numeric (string text)

```
str_var = 'hello'
num_var = int(str_var)
num_var
```

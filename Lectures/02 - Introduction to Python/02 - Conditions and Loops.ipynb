# Conditions

```
grade = 89
```

```
if grade == 89:
  print('Very Good')
```

```
if grade != 89:
  print('Not Very Good')
```

```
if grade >= 60:
  print('Success')
```

```
if grade >= 80 and grade < 90:
  print('Very Good')
```

```
if grade >= 80 or grade < 90:
  print('Very Good')
```

```
if grade >= 80 and grade < 90:
  print('Very Good')
else:
  print('Not Very Good')
```

```
if grade >= 90:
  print('Excellent')
elif grade >= 80 and grade <= 89:
  print('Very Good')
elif grade >= 70 and grade <= 79:
  print('Very Good')
elif grade >= 60 and grade <= 69:
  print('Very Good')
else:
  print('F')
```

# Loops

##### Indexed for Loop

```
for i in range(10):
  print(i)
```

```
size = range(10)

for i in size:
  print(i)
```

##### Loop on String

```
str_val = 'Hi, Mohammed. How r u ?'

for character in str_val:
  print(character)
```
